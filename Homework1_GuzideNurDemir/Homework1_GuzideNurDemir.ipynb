```python
neverMarried=resp[resp["fmarno"]==0]
neverMarried
```

```python
married=resp[resp["fmarno"] >= 1]
married
```

```python
import thinkstats2
neverm_income=thinkstats2.Hist(neverMarried["totincr"], label="Income of Never Married")
neverm_income
```

```python
import thinkstats2
married_income=thinkstats2.Hist(married["totincr"], label="Income of Married")
married_income
```

```python
width = 0.5
thinkplot.PrePlot(2)
thinkplot.Hist(neverm_income, align='right', width=width)
thinkplot.Hist(married_income, align='left', width=width)
thinkplot.Config(xlabel='Income', ylabel='Frequency')
```

```python
minimum=neverMarried["totincr"].min()
maximum=neverMarried["totincr"].max()
standard=neverMarried["totincr"].std()
variance=neverMarried["totincr"].var()
mean=neverMarried["totincr"].mean()

print("Never married people's minimum: %d, maximum: %d, standard deviation: %f, variance: %f, mean: %f income values." % (minimum,maximum,standard,variance,mean))
```

```python
minimum=married["totincr"].min()
maximum=married["totincr"].max()
standard=married["totincr"].std()
variance=married["totincr"].var()
mean=married["totincr"].mean()

print("Married people's minimum: %d, maximum: %d, standard deviation: %f, variance: %f, mean: %f income values." % (minimum,maximum,standard,variance,mean))
```
